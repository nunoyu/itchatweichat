```python
friendList = itchat.get_friends()
```

```python
import pandas as pd
import numpy as np
namelist = np.zeros((len(friendList),2))
namelist = pd.DataFrame(namelist, columns=['NickName', 'UserName'])
for idx in namelist.index:
    namelist.ix[idx,'NickName'] = friendList[idx]['NickName']
    namelist.ix[idx,'UserName'] = friendList[idx]['UserName']
namelist.to_excel('namelist.xls')    
```

```python
def findusername(NickName):
    '''
    function to find out username with nickname 
    Paramaters:
    ===========
    NickName: str
        input the NickName
    '''
    searchid = NickName 
    for i, c in enumerate(friendList):
        if c['NickName'] == searchid:
            return friendList[i]['UserName']
```
